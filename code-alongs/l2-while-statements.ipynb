## Example - birds and oil
There is an oil leakage causing the bird population in an island to 1/2  in each year. From start there were 80 000 bords. 
How many years does it take for it to have 1/10 remaining?

````python
birds = 80000
years = 0

while birds >= 8000:
    birds = birds / 2
    years += 1
   
    print(f"Year {years}, the population will have dropped to {birds}")


````

## Example - Milk
* 1L of milk has 1 500 000 bacterias in room temperature
* Bacterias increases with 50% each hour in room temp
* Milk get soured after 10 000 000 bacterias
* How many hours does it take til milk sours(whole hours)

````python
import matplotlib.pyplot as plt


bacterias = 1.5e6    #- 1.5 * 10^6
factor = 1.5
hours = 0
sour = 1e7  # 10^7
# 1.5 increase
# soured when 10 000 000 is reached

bacterias_list = [bacterias]
print(bacterias_list)


while bacterias < sour:
    bacterias *= factor
    hours += 1
    bacterias_list.append(bacterias)

print(f"Takes {hours}h for milk to get soured")
print (bacterias_list)

plt.plot(bacterias_list, "o-")
plt.xlabel("Hours")
plt.ylabel("Bacterias")
plt.title("Bacterias in Milk")
plt.grid();
````
