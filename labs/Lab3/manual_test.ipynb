### Rectangle

```python
rectangle = Rectangle(x=0,y=0,side1=2, side2=2)
print(rectangle)
print(f"-"*50)

print(f"Rectangle Area: {rectangle.area()}")
print(f"Rectangle Perimeter: {rectangle.perimeter()}")
print(f"-"*50)

rectangle1 = Rectangle(x=0,y=0,side1=2, side2=1)
rectangle2 = Rectangle(x=0,y=0,side1=1, side2=2)
print(f"Shape 1: {rectangle1}")
print(f"Shape 2: {rectangle2}")
print(f"Is shape 1 equal to shape 2?: {rectangle1 == rectangle2}")
print(f"-"*50)
```

### Rectangle Plots

```python
rectangle = Rectangle(x=0,y=0,side1=2, side2=2)
print(f"-"*25 + "PLOTS" + "-"*25)
print(f"Shape with equal sides(2, 2). Should look like an square")
rectangle.plot_rectangle()
print(f"-"*50)

rectangle = Rectangle(x=0,y=0,side1=2, side2=1)
print(f"Shape with uneven sides(2, 1). Should look like an rectangle")
rectangle.plot_rectangle()
print(f"-"*50)


rectangle = Rectangle(x=0,y=0,side1=2, side2=1)
print(f"Rectangle\nCheck if point(0.25, 0.25) is inside: {rectangle.is_inside(0.25, 0.25)}") 
rectangle.plot_rectangle(0.25, 0.25)

rectangle = Rectangle(x=0,y=0,side1=2, side2=1)
print(f"Rectangle\nCheck if point(3, 3) is inside: {rectangle.is_inside(3, 3)}") # False
rectangle.plot_rectangle(3, 3)
```
